There are 3 function for type conversion
int(): convert to integer data type
str(): convert to string data type
float(): convert to decimal

```python
# simple calculator
print("This is a simple calculator.")
# convert results of input function to integers
number1 = int(input("Enter the first number: ")) # input converted to int
number2 = int(input("Enter the second number: ")) # input converted to int

print(number1, "added to", number2, "is", number1+number2)
```

```python
# simple calculator
print("This is a simple calculator.")
# convert results of input function to integers
number1 = float(input("Enter the first number: ")) # input converted to float
number2 = float(input("Enter the second number: ")) # input converted to float
# float type conversion is safer. Because int function cant convert float to integer

print(number1, "added to", number2, "is", number1+number2)
```
